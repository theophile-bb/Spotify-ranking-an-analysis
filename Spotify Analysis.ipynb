## Data cleaning

A process of data cleaning has to be done to make the data usable for our study. We start of by removing the columns that won't be in any use :

```python
df2 = df.drop(columns = ["Unnamed: 0","uri","artist_img", "album_cover", "collab", "pivot","artist_id","album_num_tracks"])
df2
```

We then have to change the types of our columns. Here all the columns have the type 'object' so we're gonna change the types to better fit the data type in each case :

```python
df2=df2.astype(str)
df2.dtypes
```

```python
df2 = df2[df2['rank'] != 'rank']
df2.reset_index(drop=True, inplace=True)
df2
```

```python
def fix_dates(x):
    if len(str(x)) == 4:  # If it's a year like '2005'
        return f"{x}-01-01"
    return x

df2['release_date'] = df2['release_date'].apply(fix_dates)
df2['release_date'] = pd.to_datetime(df2['release_date'], errors='coerce')

df2['rank']=df2['rank'].astype({'rank': int})
df2['artists_num']=df2['artists_num'].astype({'artists_num': float})
df2['peak_rank']=df2['peak_rank'].astype({'peak_rank': int})
df2['previous_rank']=df2['previous_rank'].astype({'previous_rank': int})
df2['weeks_on_chart']=df2['weeks_on_chart'].astype({'weeks_on_chart': int})
df2['streams']=df2['streams'].astype({'streams': int})
df2['mode']=df2['mode'].astype({'mode': float})
df2['danceability']=df2['danceability'].astype({'danceability': float})
df2['energy']=df2['energy'].astype({'energy': float})
df2['key']=df2['key'].astype({'key': float})
df2['loudness']=df2['loudness'].astype({'loudness': float})
df2['speechiness']=df2['speechiness'].astype({'speechiness': float})
df2['acousticness']=df2['acousticness'].astype({'acousticness': float})
df2['liveness']=df2['liveness'].astype({'liveness': float})
df2['instrumentalness']=df2['instrumentalness'].astype({'instrumentalness': float})
df2['valence']=df2['valence'].astype({'valence': float})
df2['tempo']=df2['tempo'].astype({'tempo': float})
df2['duration']=df2['duration'].astype({'duration': float})
df2['week'] = pd.to_datetime(df2['week'])
```

```python
df2.dtypes
```

After changing the types, we check if our rows contain 'NaN' values, and if so, we fill these NaN with the mean of the column.

```python
isnull = df2.isnull().sum()/len(df2)

print("NaN values:\n", isnull)
```

```python
mean_danceability = df2['danceability'].mean()
mean_energy = df2['energy'].mean()
mean_key = df2['key'].mean()
mean_mode = df2['mode'].mean()
mean_loudness = df2['loudness'].mean()
mean_speechiness = df2['speechiness'].mean()
mean_acousticness = df2['acousticness'].mean()
mean_instrumentalness = df2['instrumentalness'].mean()
mean_liveness = df2['liveness'].mean()
mean_valence = df2['valence'].mean()
mean_tempo = df2['tempo'].mean()
mean_duration = df2['duration'].mean()

df2['danceability'].fillna(mean_danceability, inplace=True)
df2['energy'].fillna(mean_energy, inplace=True)
df2['key'].fillna(mean_key, inplace=True)
df2['mode'].fillna(mean_mode, inplace=True)
df2['loudness'].fillna(mean_loudness, inplace=True)
df2['speechiness'].fillna(mean_speechiness, inplace=True)
df2['acousticness'].fillna(mean_acousticness, inplace=True)
df2['instrumentalness'].fillna(mean_instrumentalness, inplace=True)
df2['liveness'].fillna(mean_liveness, inplace=True)
df2['valence'].fillna(mean_valence, inplace=True)
df2['tempo'].fillna(mean_tempo, inplace=True)
df2['duration'].fillna(mean_duration, inplace=True)
```

```python
df2
```

## Data visualizations

#### In this part we are gonna make many visualizations based on our data to highlight its specificities.

### Genre distribution

First, here is a little overview of the distribution of the genre of the tracks in the dataset. We can see that it is mostly pop, rap and latino music such as reggaeton.

```python
genre_distribution = df2['artist_genre'].value_counts().head(15)

fig = px.pie(genre_distribution, values=genre_distribution, names=genre_distribution.index, title='Genre Distribution in Weekly Top Songs')
fig.show()
```

### Streams by region of the world

Let's see the sreams total classified by region. We can see that the most famous artists that are grouped in the 'global' region (as their renoun is at the scale of the world) are way ahead in terms of streams. It is also due to the data temporal range not being the same: for  the 'Global' charts, there is data from the week of 2016/12/29 ~ 2022/07/14 whereas for the other dataset the data is gathered for week of 2021/02/04 ~ 2022/07/14. The first real region when it comes to streams is Europe, followed by America.

```python
region_popularity = df2.groupby('region')['streams'].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
plt.bar(region_popularity.index, region_popularity.values)
plt.xlabel('Region')
plt.ylabel('streams count')
plt.title('streams count of Weekly Top Songs by region')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
region_distribution = region_popularity

fig = px.pie(region_distribution, values=region_distribution, names=region_distribution.index, title='Popularity in number of streams around the regions of the world')
fig.show()
```

### How many songs does an average user listens to ?  

The data we use is the Spotify data from the 2021/02/04 to the 2022/07/14. It is equivalent to 75 weeks in total. We're gonna use all the streams on this time span without the data from the global chart to estimate the number of songs from this spotify top an average user listens to.

```python
nb_streams = df2[~df2["country"].isin(["Global", "country"])]
nb_streams = nb_streams["streams"].sum()
print('Total number of tracks listened on the duration : ', nb_streams)
```

This statistic study gives us the information that there are surely around 500 millions active spotify users in the world in 2023 :
https://fr.statista.com/statistiques/574665/spotify-nombre-d-utilisateurs-actifs-dans-le-monde/ .
Thanks to this, we can deduce the average number of 'popular' tracks from the spotify top an average user listens to.

```python
weeks = nb_streams / 500000000
weeks
```

An average spotify user listens to (or at least plays) 845 songs.

```python
weekly = weeks/75
weekly
```

That makes around 11 songs from this top played by an average spotify user during a week.

### Streams by country

After focusing on the streams by regions, we can go more into detail and focus on the number of streams by country. We can clearly see that the USA are the first country for streams.

```python
country_popularity = df2.groupby('country')['streams'].sum().sort_values(ascending=False)

top_20_countries = country_popularity.head(21)

# Remove country = Global
top_20_countries2 = top_20_countries[1:]
plt.figure(figsize=(10, 6))
plt.bar(top_20_countries2.index, top_20_countries2.values)
plt.xlabel('Country')
plt.ylabel('streams count')
plt.title('streams count of Weekly Top Songs by country (20 first)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The problem we with our dataset is that the first value of the country top is 'global' to reffer to world famous artists. Unfortunately we had to remove this value to facilitate our following study because it doesn't represent any country and is greatly superior to our second entry.

```python
country_popularity = df2.groupby('country')['streams'].sum().sort_values(ascending=False)

pop_df = pd.DataFrame(country_popularity)
pop_df['country'] = pop_df.index
pop_df = pop_df.reset_index(drop=True)
pop_df = pop_df[['country','streams']]
pop_df = pop_df[~pop_df["country"].isin(["Global", "country"])]
pop_df
```

Here is a visualization of these numbers on a worldmap :

And finally here is a comparison of these two plots :

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=top_artists_solo.index,
                y=top_artists_solo.values,
                name='solo',
                marker_color='rgb(55, 83, 109)'
                ))
fig.add_trace(go.Bar(x=top_artists_feat.index,
                y=top_artists_feat.values,
                name='feat',
                marker_color='lightsalmon'
                ))
fig.update_layout(barmode='group', xaxis_tickangle=-45)

fig.show()
```

Not that many artists are both in the top 20 for solo and featuring songs. Ed Sheeran and The Weeknd have much more streams on their solo songs than their featurings. In the other hand Dua Lipa, Post Malone and Drake have more strems son their featuring songs. The last artist is Bad Bunny, who also has much much more streams for his featuring songs than his solo songs (twice as much).

Let's compare the 2 tops with the total number of streams :

```python
featstreams = df3['streams'].sum()
featstreams
```

```python
solostreams = df4['streams'].sum()
solostreams
```

We can clearly see that the songs with featurings generate twice as much streams than the solo songs. We can deduce that the songs in featuring are much more popular than solo songs. It is one of the criterias that can explain the popularity of a song. We think that it can be explained by the fact that a song featuring multiple artists will appeal to the fan bases of all the artists participating, therefore generating much more streams.

### Weeks on chart

We'll now see the songs and the artists that stay in the chart the longest, based on the number of weeks.

```python
weeks = df2[['artist_individual','weeks_on_chart']]
weeks = weeks.groupby('artist_individual')['weeks_on_chart'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6))
plt.bar(weeks.index, weeks.values)
plt.xlabel('Artists')
plt.ylabel('number of total weeks cumulated among all songs')
plt.title('Artists with the most cumulated weeks in the top')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Once again, we can clearly see that Bad Bunny is far ahead in the ranking because of his numerous collaborations and featurings. Even if the top singers are significantly the same as the streams, the ranking is however different. For example The Weeknd only comes 7th when it comes to total streams, but he comes 2nd here as his music is widely enjoyed and he has a pretty loyal fanbase.
